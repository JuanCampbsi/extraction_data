## Expressões CRON

São expressões que nos permitem definir intervalos de execução de determinadas tarefas, as mesmas são definidas pela seguinte sintaxe:

```
* * * * * * *
| | | | | | |
| | | | | | +_ _ _ _  Ano (Exemplo 2023)
| | | | | +_ _ _ _ _  Dia da semana (intervalo 1 - 7)
| | | | +_ _ _ _ _ _  Mês (Intervalo 1 -12) 
| | | +_ _ _ _ _ _ _  Dia do Mês (Intervalo 1 - 31)
| | +_ _ _ _ _ _ _ _  Hora (Intervalo 0-23)
| +_ _ _ _ _ _ _ _ _  Minuto (Intervalo 0 - 59)
+_ _ _ _ _ _ _ _ _ _  Segundo (Intervalo 0 - 59)
```

Alguns exemplos Simples de Expressões CRON:

0 10 * * * * * - Nos remete a uma execução no segundo zero do minuto 10 de toda hora, dia, dia da semana, mes, ano

## Enriquecendo nossas Expressões CRON

Alguns caracteres nos permitem trazer uma maior flexibilidade para nossas cron expressions, vamos ver agora, alguns destes caracteres:

- "*" - Referencia a todos, por exemplo: Toda hora, todo Minuto, todo mês
- "?" - Referencia qualquer, geralmente usado para inicio de intervalos de execução ex: 0 ?/10 * * * * * - A cada 10 minutos a partir de qualquer minuto (Alguns agendadores acabam por não aceitar)
- "-" - Define um intervalo de ocorrência ex: 0 0 0 * * 1-7 * - Executar a meia noite de segunda a sexta
- "," - Lista de Valores ex: 0 0 0 * * 1,3,5 * - Executar toda segunda quarta e sexta
- "/" - Valores Incrementais, por exemplo: * 0/10 * * * * * - Executar a cada 10 min a partir do minuto zero

## Arquitetura de Eventos com Webhooks

### **O que são Webhooks?**

Webhooks são interfaces que permitem a subscrição de um serviço em um determinado servidor para que o servidor possa enviar eventos aos subscritor sempre que houver um novo evento

**Diferentemente das APIs** comuns, onde se disponibiliza um endpoint para um client X consumí-lo, no caso dos Webhooks, nós implementamos um endpoint de acordo com os padrões do serviço ao qual desejamos nos inscrever e depois nos inscrevemos neste serviço para que sempre que houver um novo evento, possamos receber esse evento em nosso ambiente.
