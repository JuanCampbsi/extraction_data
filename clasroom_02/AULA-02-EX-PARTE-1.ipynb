```python
import json

response_products = requests.get(url="https://fakestoreapi.com/products")
response_categories = requests.get(url="https://fakestoreapi.com/products/categories")

print(f"Products Reponse Através do método JSON: {response_products.json()}")
print(f"Categories Reponse Através do método JSON: {response_categories.json()}")

print(f"Products Reponse através do json.loads: {json.loads(response_products.text)}")
print(f"Categories Reponse através do json.loads: {json.loads(response_categories.text)}")
```

2 - Implemente o código anterior utilizando a estratégia de retries vista em sala de aula


```python
def gerencia_requisicao(url: str, endpoint: str, max_retries: int, current_retry: int = 0):
    
    response = requests.get(url=f"{url}/{endpoint}")
    
    if (response.status_code >= 400 and response.status_code <= 499) or (response.status_code >= 500 and response.status_code <= 599):
        if current_retry > max_retries:
            return gerencia_requisicao(
                url=url,
                endpoint=endpoint,
                max_retries=max_retries,
                current_retry=current_retry+1
            )
        else:
            print(f"Erro ao processar requisição mesmo após {current_retry} tentativa(s). Código de Erro: {response.status_code}")
    
    
    return response


products_response_com_retry = gerencia_requisicao(
    url="https://fakestoreapi.com", 
    endpoint="products", 
    max_retries=3
)

categories_response_com_retry = gerencia_requisicao(
    url="https://fakestoreapi.com", 
    endpoint="products/categories", 
    max_retries=3
)

print(f"Categories {categories_response_com_retry.text}")
print(f"Products {products_response_com_retry.text}")
```
